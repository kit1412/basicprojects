```
import random
from collections import deque

item = deque(["wjiw", 2333, 4556, "ejfiwhgw", 0])


print("\nThe list as of now: ", item)
#printing the list of values

x = int(input("\nEnter a random integer\n"))

#so that a different value is produced each time
l = ["wihfi", -78, 665, 3.5, -8888, 0, 22200, "fwjihfhw"]
#chooses a value from the list above depending on x, thus offering some amount of randomness

random.seed(x)
item.appendleft(random.choice(l))

random.seed(x % 10)
item.appendleft(random.choice(l))

random.seed(x*10)
item.append(random.choice(l))

random.seed(x/10)
item.append(random.choice(l))

print("\nThe list as of now: ", item)
#printing the list of values
```

list 1. ME, ISE, CSE, MATH

list 3. 17, 18, 18, 17, 19, 18
(b) Print Sruti by accessing through index
(c) find maximum from list 3 of the nested list

```
info = [["ME", "ISE", "CSE", "MATH"], ["Ram", "Shyam", "Sruti", "Pankaj", "Anuja", "Anjali"], [17, 18, 18, 17, 19, 18]]
#defining the list as required

print(info[1][2])

max = info[2][0]
#setting the first mark as maximum

#to store the index of all the marks which are equal


for i in range(0, 6):
  if(info[2][i] >= max):
    max = info[2][i]

#finding the largest element
    
index = set()
for i in range(0, 6):
  if (max == info[2][i]):
    index.add(i)
#checking for repetitions


print("The largest element is ", max, "and it is present at index ", index)
#printing all the indices of the largest element
```

3. Suppose people of list2 in question #2 went to watch a film in a theatre. There are certain
rules in the theatre regarding selling of tickets. The rules are as follows -
i.Everyone has to buy their own tickets.
ii. In front of the counter only 4 persons can stand at a time in a queue.
iii. Confirm every person gets a ticket following Fist in First Out (FIFO) condition.
Implement these rules and display the entry and exists from the queue till everyone gets their
tickets.

```
names = ["Ram", "Shyam", "Sruti", "Pankaj", "Anuja", "Anjali"]

queue = []

for i in range(0, 4):
  queue.append(names[i])
  print(queue[i], "stood in the queue")

n = len(names)
#obtaining the length of the list

for i in range(0, n - 4):
  print(queue.pop(0), "took the ticket and left the queue")
  print(names[i + 4], "stood in the queue")
  queue.append(names[i + 4])
#between the first four and the last four

for i in range(0, 4):
  print(queue.pop(0), "took the ticket and left the queue")
#for the last four people
```
