```
class Accounts:

 def __init__(self, n, add):
  self.name = n
  self.address = add
  self.amount = 0

 def deposit(self, amt):
  self.amount += amt
#adding the amount deposited  

 def withdraw(self, amt):
  self.amount -= amt
#subtracting the amount withdrawn

 def balance(self):
  print(self.amount)
#printing the balance

 def interest(self, time):
  pass
#creating an empty function as asked
```


2.      Create another class for savings account.  Inherit interest()  method to calculate the interest amount for a customer. Assume the rate of interest is 20% per year.


```
class SavingsAccount (Accounts):
 
  def interest(self, n):
    #simple interest
    
    self.interest = self.amount * .2 * n  
    print("Interest = ", self.interest)
```
